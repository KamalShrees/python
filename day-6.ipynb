```python
src = "H:\\IMAGES\\ChisaPani"
os.chdir(src)                                 
modify=os.listdir(src)                      

valid_ext=['mp4','3gp']
if not os.path.exists(src+'\Videos'):
    os.mkdir('Videos')
for i,file in enumerate(modify):
    ext=file.split('.')[-1]
    if ext.lower() in valid_ext:
        os.rename(file,f'Chisapani_{i}.{ext}')  #renames function can also be used to rename and move file
        shutil.move(src+ f'\\Chisapani_{i}.{ext}',src+'\Videos')
```
